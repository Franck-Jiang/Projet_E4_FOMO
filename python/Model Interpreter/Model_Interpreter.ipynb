```python
import subprocess

def check_docker_status():
    try:
        # Vérifier si le service Docker est en cours d'exécution
        docker_service_status = subprocess.run(["docker", "info"], capture_output=True, text=True)
        if "Server:" not in docker_service_status.stdout:
            print("Le service Docker n'est pas en cours d'exécution.")
            return False
        else:
            print("Le service Docker est en cours d'exécution.")
            return True
    except FileNotFoundError:
        print("Docker n'est pas installé sur votre système.")
        return False

def check_docker_pull():
    try:
        # Vérifier si la commande docker pull fonctionne
        docker_pull_status = subprocess.run(["docker", "pull", "hello-world"], capture_output=True, text=True)
        if "Status: Downloaded newer image for hello-world" in docker_pull_status.stdout:
            print("La commande 'docker pull' fonctionne correctement.")
            return True
        else:
            print("Erreur lors de l'exécution de la commande 'docker pull'.")
            return False
    except FileNotFoundError:
        print("Docker n'est pas installé sur votre système.")
        return False

def main():
    # Vérifier l'état du service Docker
    docker_status = check_docker_status()
    if not docker_status:
        return

    # Vérifier la commande docker pull
    docker_pull = check_docker_pull()
    if not docker_pull:
        return

    print("Toutes les vérifications ont été effectuées avec succès.")

if __name__ == "__main__":
    main()
```

```python
import os
import subprocess

# Chemin vers le dossier contenant les fichiers C
c_files_dir = r"C:\Users\remil\Desktop\Model_Interpreter\Projet_E4_FOMO\c"

# Parcourir les noms des fonctions
for function_name in function_names:
    # Construire le chemin du fichier C
    c_file_path = os.path.join(c_files_dir, f"{function_name}.c")

    # Vérifier si le fichier C existe
    if os.path.exists(c_file_path):
        # Compiler le fichier C
        compile_command = ["gcc", c_file_path, "-o", f"{function_name}.exe"]
        compilation_result = subprocess.run(compile_command, capture_output=True, text=True)
        
        if compilation_result.returncode == 0:
            print(f"Compilation réussie pour la fonction {function_name}")
            
            # Exécuter le fichier compilé
            execution_command = [f"{function_name}.exe"]
            execution_result = subprocess.run(execution_command, capture_output=True, text=True)
            
            print(f"Résultat de l'exécution pour la fonction {function_name}:")
            print(execution_result.stdout)
        else:
            print(f"Échec de la compilation pour la fonction {function_name}:")
            print(compilation_result.stderr)
    else:
        print(f"Le fichier C pour la fonction {function_name} n'a pas été trouvé.")
```
