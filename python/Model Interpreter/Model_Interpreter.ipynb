```python
import json

# Lis l'interpreteur au format .json
file_path = r"C:\Users\remil\Desktop\Model_Interpreter\ei-beer-vs.-cans-fomo-object-detection-tensorflow-lite-int8-quantized-model.json"
try:
    with open(file_path, "r") as json_file:
        data = json.load(json_file)
except FileNotFoundError:
    print("File not found:", file_path)
    exit()

# On récupère les noms des fonctions
if "subgraphs" in data:
    functions = []
    for subgraph in data.get("subgraphs", []):
        operators = subgraph.get("operators", [])
        for operator in operators:
            function = {}
            builtin_options_type = operator.get("builtin_options_type", "")
            if builtin_options_type:
                function["name"] = builtin_options_type.replace("Options", "")
                function["parameters"] = operator.get("builtin_options", {})
                functions.append(function)

    print("Liste des fonctions avec leurs paramètres :")
    for func in functions:
        print("Nom de la fonction :", func["name"])
        print("Paramètres :", func["parameters"])
        print()

# Chemin du fichier de sortie texte
output_file_path = "functions_parameters.txt"

# Écriture des informations dans le fichier texte
with open(output_file_path, "w") as txt_file:
    txt_file.write("Liste des fonctions avec leurs paramètres :\n")
    for func in functions:
        txt_file.write("Nom de la fonction : {}\n".format(func["name"]))
        txt_file.write("Paramètres :\n")
        for param, value in func["parameters"].items():
            txt_file.write("\t{} : {}\n".format(param, value))
        txt_file.write("\n")

print("Les informations ont été écrites dans le fichier :", output_file_path)
```
